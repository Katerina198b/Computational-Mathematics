# Depedency of processing time on threads numper

```python
import numpy as np
from matplotlib import pyplot as plt
%pylab inline
```

## MPI LINEAR

```python
#kernels = r
time = [0.135312, 0.072150, 0.051348, 0.044295, 0.041477, 0.040574, 0.038920, 0.037534]
```

```python
plt.plot(range(1,9), time, 'r')
plt.ylabel('Time', fontsize = 'x-large')
plt.xlabel('Kernels', fontsize = 'x-large')
```
